#### **Array indexing**

```python
x = np.array([1,3,5,7,9])
x
print(f"The first element in the array is {x[0]}.") #first element in the array
print(f"The last element in the array is {x[-1]}.") #last element in the array
print(f"The second to fourth element is {x[1:4]}") #to access the second to fourth elements
print(f"The element between the second and fourth after 2 step is {x[1:4:2]}")
```

```python
#for 2 dimensional arrays
y = np.array([[1,3,5,7,9], [2,4,6,8,10], [3,6,9,12,15]])
print(f" The two dimensional array is: \n {y}")
```

```python
print(y[0,1]) #row 1, column 2 mathematically
print(y[1,3]) #row 2, column 4 mathematically
print(y[2,4]) #row 3, column 5 mathematically
```

```python
print(y[0,1:4]) #row 1, column 2, uptill item before index number 4 mathematically
```

```python
print(f"For this, we have \n{y[:, 0:3]}") #all the rows, but column 1 to 3 mathematically
```

```python
print(y[0:, 2:]) #everything from row 0 to 2, and also column 2 to the end. That is, where row 1 column 3 starts, and row 3 column 5 ends mathematically
```

```python
print(y[0:, 2:4]) #here row 1 and column 3 but also ends with row 3, column 4
```

```python
print(y[1:, 2:4]) #to get 6,8,9,12
print()

print(y[:, 0:3]) #all rows, first to third column
print()

print(y[1:, :])
```

#### **Reshaping arrays**

You can only reshape to the size of your initial array. For example, you can't reshape an array with a size of 10 to a size of 8.

To do this, call the reshape method and pass the (row, column argurment)

```python
z = np.arange(4)
print(f"{z} \n")

new_z = z.reshape(2,2)
print(f"The  array of new_z is: \n {new_z}")
print()

#to reshape a 2D to another 2D
arr = np.array([[1,2,3],[4,5,6]])
print(f"The original 2d array is: \n {arr}")
print(f"The shape of this array is {arr.shape} \n\n")


new_arr = arr.reshape(3,2)
print(f"The reshaped 2d array is: \n {new_arr}")
print(new_arr)
print(f"The shape of this array is {new_arr.shape} \n")
```

#### **Concatenation & splitting arrays**

Row wise concatenation must have the same number of columns so that they will be stacked on each other.

To split an array, know the index where the splitting should occur.

```python
#to concatenate arrays
a = np.array([1,2,3])
b = np.array([4,5,6])
c = np.concatenate((a,b)) #pass into a list but mine worked in a tuple
print(c)
```

```python
y
```

```python
y_y = np.concatenate((y,y)) #row_wise
print(y_y)
print()
#to stack as new columns, you'll pass in an arguement
yy_y = np.concatenate((y,y), axis=1) #column_wise
print(yy_y)
```

```python
#another way is to use H stack and V stack method
#using H-stack
y_hstack = np.hstack((y,y))
print(y_hstack)
print()
#using V-stack
y_vstack = np.vstack((y,y))
print(y_vstack)
```

```python
#to split an array
g = np.arange(15, 30)
print(g)
print()

h,i,j = np.split(g, [5,10]) #i starts from index 5, j starts from index 10 for 3 arrays
k,l = np.split(g, [10]) #i starts from index 5, j starts from index 10 for 2 arrays
print(h)
print(i)
print(j)
print(k)
print(l)
```
