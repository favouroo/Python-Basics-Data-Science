#### **With the try-except block**

```python
try:
  first_num = float(input("Enter a number: "))
  sec_num = float(input("Enter a number: "))
  result = first_num/sec_num
except ZeroDivisionError:
  print(f"You can't divide {first_num} or any number by zero.")
except ValueError:
  print("You didn't enter a valid number.")
else:
  print(f"{first_num_1} divided by {sec_num_1} is {result}.")
```
