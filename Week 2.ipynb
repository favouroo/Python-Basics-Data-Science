### Sort Method

```python
#Sorting items in a list

unordered_list = [20, -1, 30, 70, 199, 83]
print (unordered_list)

unordered_list.sort() #Ascending order
print (unordered_list)

unordered_list.sort(reverse = True) #Descending order
print (unordered_list)

print(unordered_list[:3]) #Top 3 numbers
print(unordered_list[-3:]) #Least 3 numbers

#Limitation: Sort method only works when you want to alter the original list. If you store in a new variable, you'll get none.
```

```python
#Limitation: Sort method only works when you want to alter the original list. If you store in a new variable, you'll get none.
#using the SORTED function

new_unordered_list = [20, -1, 30, 70, 199, 83]
ordered_list = sorted(unordered_list) #solves the limitation of the sort method
print (ordered_list)

descending_order = sorted(new_unordered_list, reverse = True)
print (descending_order)
```

###Reversing a list

```python
#Call the reverse method
some_unordered_list = [20, -1, 30, 70, 199, 83]
print(some_unordered_list)
some_unordered_list.reverse() #doesnt sort, just reverses position
```

###Copying a list

```python
number_list = [10, 20, 30, 40, 50, 60, -1, 0, 1000]
print(number_list)

copied_num_list = number_list.copy()
copied_num_list[-2] = 18 #access element first by indexing
print(copied_num_list)
```

```python
#to get the index of a particular element in a list, use the index method
print(number_list.index(0))
print(number_list.index(30))
print(number_list.index(60))
sum(number_list)
print(min(number_list))
print(max(number_list))
```

##Tuples
Use tuples when you just want to store variables you dont want to change. A TUPLE can also contain different data types. You can either use parenthesis or not, as long as you separate the item wwith a comma.

```python
number_tuple = (10, 30, 55, 66,26, 5, 34, 23, 77)
print(number_tuple)
print(type(number_tuple))
print()
numberr_tuple = 11, 82, 32, 24, 56, 74, 38, 32, 22, 9, -6
print(numberr_tuple)
type(numberr_tuple)
```

```python
print(len(number_tuple))
print(len(numberr_tuple))
print()
print(number_tuple[0])
print(numberr_tuple[0])
print(number_tuple[5])
print(numberr_tuple[5])
print()
print(number_tuple[-3:])
print(numberr_tuple[-3:])
print(number_tuple[0:4])
print(numberr_tuple[0:4])
```

```python
#Tuple has only two methods; count and index
#Counts checks the occurence of a avalue
print(number_tuple.count(50))
print(number_tuple.count(66))
print()

print(numberr_tuple.count(50))
print(numberr_tuple.count(32))
print()
print(number_tuple.index(10))
print(number_tuple.index(66))
print(numberr_tuple.index(32))
print(numberr_tuple.index(38))
```

###**Dictionaries**
This is the foundation for dataframes. They come in form of key-value pairs. Dicitonaries are unordered.
Keys are unique identifiers for a particular element.
Dictionaries are wrrapped in currly brackets.

```python
student_info = {"name":"Favour Okolie", "age":20, "course":"Engineering for Environmental Sustainability and International Cooperation (EESIC, MEng)"}
print(student_info)
print(type(student_info))
print()

#accessing elements in a dictionary
print(student_info["course"])
print(student_info.keys()) #to get the keys
print(student_info.values()) #to get the values
print(student_info.items()) #to get key-value pairs

print()
student_info["age"] = 25 #dictionaries are mutable. To change a value, do this
print(student_info)
```
